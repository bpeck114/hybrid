# Hybrid: Radius Study
`20km_study1A.ipynb`
: The goal of this jupyter notebook is to produce plots of point-spread function (PSF) metrics and RMS wave-front error (WFE) for the on-axis hybrid LGS trade study. The current plots include:
* Height vs. Strehl
* Height vs. FWHM
* Height vs. EE80
* Height vs. WFE
* Height vs. WFE (1000 step burn-in)

## Setup

````python
import os
import importlib
import pylab as plt
import numpy as np


# Specific packages (may need extra steps to import)
from paarti.utils import maos_utils
from paarti import psfs, psf_plots
from paarti.psf_metrics import metrics
from bpeck.hybrid import on_axis_study
````

````python
radius = [0, 1, 2, 5, 7, 10, 60] # arc seconds
seed = 1
bandpass = 3
start_index = 1000

%matplotlib inline
````

## Results

### Height vs. Strehl

````python
strehl_800 = np.array([strehl_0[0], strehl_1[0], strehl_2[0], strehl_5[0], strehl_7[0], strehl_10[0], strehl_60[0]])
strehl_1000 = np.array([strehl_0[1], strehl_1[1], strehl_2[1], strehl_5[1], strehl_7[1], strehl_10[1], strehl_60[1]])
strehl_1250 = np.array([strehl_0[2], strehl_1[2], strehl_2[2], strehl_5[2], strehl_7[2], strehl_10[2], strehl_60[2]])
strehl_1650 = np.array([strehl_0[3], strehl_1[3], strehl_2[3], strehl_5[3], strehl_7[3], strehl_10[3], strehl_60[3]])
strehl_2120 = np.array([strehl_0[4], strehl_1[4], strehl_2[4], strehl_5[4], strehl_7[4], strehl_10[4], strehl_60[4]])

plt.figure(figsize=(4,4))
plt.plot(radius, strehl_2120, color = "red", label = "2120 nm")
plt.plot(radius, strehl_1650, color = "orange", label = "1000 nm")
plt.plot(radius, strehl_1250, color = "green", label = "1250 nm")
plt.plot(radius, strehl_1000, color = "cyan", label = "1650 nm")
plt.plot(radius, strehl_800, color = "blue", label = "800 nm")
plt.title('Radius of Rayleigh Beacon (") vs. Strehl Ratio')
plt.xlabel('Radius of Rayleigh Beacon (")')
plt.ylabel("Strehl Ratio")
plt.legend(bbox_to_anchor=(1.45, 1), loc='upper right')
plt.show()
````

### Height vs. WFE

````python
total_wfe = np.array([clos_mean_nm_0[0], clos_mean_nm_1[0], clos_mean_nm_2[0], clos_mean_nm_5[0], clos_mean_nm_7[0], clos_mean_nm_10[0], clos_mean_nm_60[0]])
tt_wfe = np.array([clos_mean_nm_0[1], clos_mean_nm_1[1], clos_mean_nm_2[1], clos_mean_nm_5[1], clos_mean_nm_7[1], clos_mean_nm_10[1], clos_mean_nm_60[1]])
ho_wfe = np.array([clos_mean_nm_0[2], clos_mean_nm_1[2], clos_mean_nm_2[2], clos_mean_nm_5[2], clos_mean_nm_7[2], clos_mean_nm_10[2], clos_mean_nm_60[2]])


plt.figure(figsize=(4,4))
plt.plot(radius, total_wfe, color = "blue", label = "Total RMS WFE")
plt.plot(radius, ho_wfe, color = "red", label = "High-Order RMS WFE")
plt.plot(radius, tt_wfe, color = "green", label = "Tip-Tilt RMS WFE")
plt.title('Radius of Rayleigh Beacon (") vs. WFE')
plt.xlabel('Radius of Rayleigh Beacon (")')
plt.ylabel("WFE")
plt.legend(bbox_to_anchor=(1.75, 1), loc='upper right')
plt.show()
````
