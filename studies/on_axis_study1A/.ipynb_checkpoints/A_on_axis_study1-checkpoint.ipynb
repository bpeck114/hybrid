# Hybrid: On-Axis Study
`A_on_axis_study1.ipynb`
: The goal of this jupyter notebook is to produce plots of point-spread function (PSF) metrics and RMS wave-front error (WFE) for the on-axis hybrid LGS trade study. The current plots include:
* Height vs. Strehl
* Height vs. FWHM
* Height vs. EE80
* Height vs. WFE

## Results

### Height vs. Strehl

````python
strehl_800 = np.array([strehl_1[0], strehl_5[0], strehl_10[0], strehl_15[0], strehl_20[0], strehl_25[0], strehl_30[0]])
strehl_1000 = np.array([strehl_1[0], strehl_5[1], strehl_10[1], strehl_15[1], strehl_20[1], strehl_25[1], strehl_30[1]])
strehl_1250 = np.array([strehl_1[0], strehl_5[2], strehl_10[2], strehl_15[2], strehl_20[2], strehl_25[2], strehl_30[2]])
strehl_1650 = np.array([strehl_1[0], strehl_5[3], strehl_10[3], strehl_15[3], strehl_20[3], strehl_25[3], strehl_30[3]])
strehl_2120 = np.array([strehl_1[0], strehl_5[4], strehl_10[4], strehl_15[4], strehl_20[4], strehl_25[4], strehl_30[4]])

plt.figure(figsize=(4,4))
plt.plot(height, strehl_2120, color = "red", label = "2120 nm")
plt.plot(height, strehl_1650, color = "orange", label = "1000 nm")
plt.plot(height, strehl_1250, color = "green", label = "1250 nm")
plt.plot(height, strehl_1000, color = "cyan", label = "1650 nm")
plt.plot(height, strehl_800, color = "blue", label = "800 nm")
plt.title("Height of Rayleigh Beacon (km) vs. Strehl Ratio")
plt.xlabel("Height of Rayleigh Beacon (km)")
plt.ylabel("Strehl Ratio")
plt.legend(bbox_to_anchor=(1.45, 1), loc='upper right')
plt.show()
````

### Height vs. WFE

````python
total_wfe = np.array([clos_mean_nm_1[0], clos_mean_nm_5[0], clos_mean_nm_10[0], clos_mean_nm_15[0], clos_mean_nm_20[0], clos_mean_nm_25[0], clos_mean_nm_30[0]])
ho_wfe = np.array([clos_mean_nm_1[0], clos_mean_nm_5[1], clos_mean_nm_10[1], clos_mean_nm_15[1], clos_mean_nm_20[1], clos_mean_nm_25[1], clos_mean_nm_30[1]])
tt_wfe = np.array([clos_mean_nm_1[0], clos_mean_nm_5[2], clos_mean_nm_10[2], clos_mean_nm_15[2], clos_mean_nm_20[2], clos_mean_nm_25[2], clos_mean_nm_30[2]])


plt.figure(figsize=(4,4))
plt.plot(height, total_wfe, color = "blue", label = "Total RMS WFE")
plt.plot(height, ho_wfe, color = "red", label = "High-Order RMS WFE")
plt.plot(height, tt_wfe, color = "green", label = "Tip-Tilt RMS WFE")
plt.title("Height of Rayleigh Beacon (km) vs. WFE")
plt.xlabel("Height of Rayleigh Beacon (km)")
plt.ylabel("WFE")
plt.legend(bbox_to_anchor=(1.75, 1), loc='upper right')
plt.show()
````
